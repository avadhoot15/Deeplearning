```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
word_string = "India is a republic nation We are proud Indians . we are Love India . India India India . Jai Hind ."
wordcloud = WordCloud(stopwords = STOPWORDS,collocations=True)
out=wordcloud.generate(word_string)
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18)
#Plotting the Wordcloud
plt.imshow(out, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import spacy
#loading the english language small model of spacy
en = spacy.load('en_core_web_sm')
stopwords = en.Defaults.stop_words
text = " we will show how to remove stopwords using spacy library"
lst=[]
for token in text.split():
 if token.lower() not in stopwords:
   lst.append(token) 
 
#Join items in the list
print("Original text : ",text)
print("Text after removing stopwords : ",' '.join(lst))
```
